```python
import pandas as pd
import json
import geopandas

buildings_gdf = geopandas.read_file("msc_buildings.gpkg")  

buildings_gdf.info()
```

```python
full_frame = pd.DataFrame(buildings_gdf) 

full_frame.head()
```

```python
street_name = "улица Довженко" 
full_frame = pd.DataFrame(buildings_gdf)
fuul_table_street = full_frame[full_frame["addr:street"] == street_name]
fuul_table_street
```

```python
frame = pd.DataFrame(buildings_gdf, columns = ["addr:street", "addr:housenumber", "start_date"])

addr_column = frame["addr:street"]
buildings_on_streets = frame[addr_column.isin([street_name, "Сретенский бульвар"])]
buildings_on_streets
```

```python
# write buildings from street

tfile = open('dataframe_buildings.txt', 'a')
tfile.write(buildings_on_streets.to_string())
tfile.close()
```
